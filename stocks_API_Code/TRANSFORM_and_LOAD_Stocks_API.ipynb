# Transform News Data

## Load & Read CSV Files

```python
# Load CSV files

news = "../stocks_data/raw_stock_news_data.csv"

# Read files and store into Pandas DataFrame

news_df = pd.read_csv(news)
news_df
```

## Clean columns

```python
# Clean Symbol Column
news_df['symbol'] = news_df['symbol'].str.replace(r'TICKER:','')

# Clean date from UTC to regular date
news_df['date'] = pd.to_datetime(news_df['date'], unit = 's')

# Clean date to remove time 
news_df['date'] = news_df['date'].dt.strftime('%m/%d/%Y')
news_df
```

## Rearrenge columns and configure dataframe width

```python
# Rearrenge Columns

news_df = news_df[['symbol','name','date','title','publisher','source_link','summary']]
news_df

# Styler to adjust columns and center text
# https://stackoverflow.com/questions/40990700/pandas-dataframes-in-jupyter-columns-of-equal-width-and-centered

dnews = dict(selector = "th", props = [('text-aligh', 'center')])

news_df.style.set_properties(**{'width': '10em', 'text-aligh': 'center'}).set_table_styles([dnews])
```

## Save to CSV and JSON CLEAN Files

```python
# Save to CSV and JSON files

news_df.to_csv("../stocks_data/CLEAN_stock_news_data.csv", index=False, encoding='utf8')

news_df.to_json("../stocks_data/CLEAN_stock_news_data.json", orient='columns')
```

# Transform Fortune 500 Table 

## Load & Read CSV File

```python
# Load CSV files

fortune = "../stocks_data/raw_fortune500table.csv"

# Read files and store into Pandas DataFrame

fortune_df = pd.read_csv(fortune)
fortune_df
```

## Replace, rename, split and drop columns

```python
# Replace NAN values with 0

fortune_df = fortune_df.replace(np.nan,0)  # https://stackoverflow.com/questions/13295735/how-can-i-replace-all-the-nan-values-with-zeros-in-a-column-of-a-pandas-datafra?rq=1

# Rename Headquarters column
fortune_df = fortune_df.rename(columns= {'Headquarters Location': 'HQ'})

# Separate/split state of Headquarters Location 
new_fortune = fortune_df['HQ'].str.split(",", n=1, expand = True) # # new data frame with split value columns 

# Creating new columns and separating values
fortune_df['HQ City'] = new_fortune[0]
fortune_df['HQ State'] = new_fortune[1]

# Drop previous columnn and unnessary columns
fortune_df.drop(columns = ['HQ','SEC filings'], inplace = True)

# Display DF

fortune_df
```

## Save to CSV and JSON CLEAN Files

```python
# Savings to CSV and JSON files

fortune_df.to_csv("../stocks_data/CLEAN_fortune500table.csv", index=False, encoding='utf8')

fortune_df.to_json("../stocks_data/CLEAN_fortune500table.json", orient='columns')
```

# Transform Financial Statements Data

## Load & Read CSV Files

```python
# Load CSV files

balance_data = "../stocks_data/raw_balance_sheets_data.csv"
cashflow_data = "../stocks_data/raw_cashflow_sheets_data.csv"
income_data = "../stocks_data/raw_income_sheets_data.csv"

# Read files and store into Pandas DataFrame

balance_sheets_df  = pd.read_csv(balance_data)
cashflow_data_df  = pd.read_csv(cashflow_data)
income_data_df  = pd.read_csv(income_data)
```

## Display, drop, slice, re_index, replace

```python
balance_sheets_df
```

```python
# Drop unnessary column maxAge
balance_sheets_df.drop(columns = ['maxAge'], inplace = True)
cashflow_data_df.drop(columns = ['maxAge'], inplace = True)
income_data_df.drop(columns = ['maxAge'], inplace = True)

# Set index to symbol
balance_sheets_df.set_index('symbol', inplace= True)
cashflow_data_df.set_index('symbol', inplace= True)

# Display results
balance_sheets_df.head()
cashflow_data_df.head()
income_data_df.head()
```

```python
# Slice iloc every 3 rows
balance_sheets_df = balance_sheets_df.iloc[::3] #https://stackoverflow.com/questions/33634142/pandas-how-to-delete-alternate-rows

cashflow_data_df = cashflow_data_df.iloc[::3]

income_data_df = income_data_df.iloc[::3]

# Reset index
balance_sheets_df = balance_sheets_df.reset_index()

cashflow_data_df = cashflow_data_df.reset_index()

income_data_df= income_data_df.reset_index()

# Replace NAN values with 0

balance_sheets_df = balance_sheets_df.replace(np.nan,0)

cashflow_data_df = cashflow_data_df.replace(np.nan,0)

income_data_df = income_data_df.replace(np.nan,0)

# Display results
balance_sheets_df
cashflow_data_df
income_data_df
```

```python
# Rearrenge columns

new_cols_b = ['symbol','endDate'] + [c for c in balance_sheets_df.columns if c not in ['symbol','endDate']]
new_cols_c = ['symbol','endDate'] + [c for c in cashflow_data_df.columns if c not in ['symbol','endDate']]
new_cols_i = ['symbol','endDate'] + [c for c in income_data_df.columns if c not in ['symbol','endDate']]

# Re-index after rearrenging
balance_sheets_df = balance_sheets_df.reindex(columns = new_cols_b)
cashflow_data_df = cashflow_data_df.reindex(columns = new_cols_c)
income_data_df = income_data_df.reindex(columns = new_cols_i)

# Clean date from UTC to regular date
balance_sheets_df['endDate'] = pd.to_datetime(balance_sheets_df['endDate'], unit = 's')

cashflow_data_df['endDate'] = pd.to_datetime(cashflow_data_df['endDate'], unit = 's')

income_data_df['endDate'] = pd.to_datetime(income_data_df['endDate'], unit = 's')

# Clean date to remove time 
balance_sheets_df['endDate'] = balance_sheets_df['endDate'].dt.strftime('%Y/%m/%d')

cashflow_data_df['endDate'] = cashflow_data_df['endDate'].dt.strftime('%Y/%m/%d')

income_data_df['endDate'] = income_data_df['endDate'].dt.strftime('%Y/%m/%d')

# Select third column to the end and convert from object to numeric
# https://stackoverflow.com/questions/36814100/pandas-to-numeric-for-multiple-columns

for b in balance_sheets_df.iloc[:,2:]:
    balance_sheets_df[b] = pd.to_numeric(balance_sheets_df[b], errors = 'coerce')

for c in cashflow_data_df.iloc[:,2:]:
    cashflow_data_df[c] = pd.to_numeric(cashflow_data_df[c], errors = 'coerce')

for i in income_data_df.iloc[:,2:]:
    income_data_df[i] = pd.to_numeric(income_data_df[i], errors = 'coerce')
```

# Transform Historical Stocks Data - Price & Dividend

## Load & Read CSV files

```python
# Load CSV files

price = "../stocks_data/raw_price_history.csv"
dividend = "../stocks_data/raw_dividend_history.csv"

# Read files and store into Pandas DataFrame

price_df  = pd.read_csv(price)
dividend_df = pd.read_csv(dividend)
```

### Prices Data

### Drop, fill, rearrenge, and format and more

```python
# Display Prices
price_df

# Drop unnessaries columns
price_df.drop(columns = ['amount','type','data', 'numerator','denominator', 'splitRatio'], inplace = True)

# Fix date column
price_df['date'] = pd.to_datetime(price_df['date'], unit = 's')

# Rearrenge columns

new_cols_p = ['symbol','date','open','close','high','low'] + [c for c in price_df.columns if c not in ['symbol','date','open','close','high','low']]

# Re-index after rearrenging
price_df = price_df.reindex(columns = new_cols_p)

price_df
```

```python
# Clean date to remove time 
price_df['date'] = price_df['date'].dt.strftime('%Y-%m-%d')

# Sort dataframe by date

price_df = price_df.sort_values(['date'], ascending = True)

# Drop data with NAN values
price_df = price_df.dropna(axis = 0, how = 'any')

# Reset index
price_df = price_df.reset_index(drop = True)

# See bottom 30 columns
price_df.iloc[-30:]
```

### Save to CSV CLEAN Files

```python
### Save to CSV CLEAN Files
price_df.to_csv("../stocks_data/CLEAN_stock_prices.csv", index=False, encoding='utf8')
```

### Dividend Data

### Drop, fill, rearrenge, and format and more

```python
# Display data
dividend_df

# Drop unnessary columns
dividend_df.drop(columns = ['data', 'numerator','denominator', 'splitRatio'], inplace = True)

# Drop data with NAN values
dividend_df = dividend_df.dropna(axis = 0, how = 'any')

dividend_df
```

```python
# Fix date column
dividend_df['date'] = pd.to_datetime(dividend_df['date'], unit = 's')

# Clean date to remove time 
dividend_df['date'] = dividend_df['date'].dt.strftime('%Y-%m-%d')

# Rearrenge data frame
# Rearrenge columns

new_cols_d = ['symbol','date','amount','type'] + [c for c in dividend_df.columns if c not in ['symbol','date','amount','type']]

# Re-index after rearrenging
dividend_df = dividend_df.reindex(columns = new_cols_d)


# Sort dataframe by date

dividend_df  = dividend_df.sort_values(['date'], ascending = True)

# Reset index
dividend_df  = dividend_df.reset_index(drop = True)

# See bottom 30 columns
dividend_df .iloc[-30:]
```

### Save CSV Clean files

```python
### Save to CSV CLEAN Files
dividend_df.to_csv("../stocks_data/CLEAN_stock_dividend.csv", index=False, encoding='utf8')
```
